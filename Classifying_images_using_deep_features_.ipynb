# ***Main codes***

```
import turicreate as tc
```

## *Loading train and test data*

```
image_train = tc.SFrame("image_train_data")
image_test = tc.SFrame("image_test_data")
```

### *Visualising Data*

```
image_train.explore()
```

```
image_test.explore()
```

## *Train an image classifier on raw image pixels(Not recommanded)*

```
raw_image_model = tc.logistic_classifier.create(image_train,features = ['image_array'],target = 'label')
```

### *Making some prediction*

```
image_test[0:5]['image', 'label'].explore()
```

```
raw_image_model.predict(image_test[0:5])
```

So we can see it is a very bad model .<br>
Our simple image pixel classifier model only recognise 1 image out of 5 😰

```
raw_image_model.evaluate(image_test)
```

<strong>'accuracy': 0.48275</strong>
<br>i.e. very bad model



## *Train image classifier using deep features*

```
len(image_train)
```

```
#deep_learning_model = turicreate.load_model('imagenet_model_iter45')
#image_train['deep_features'] = deep_learning_model.extract_features(image_train)
```

```
image_train
```

```
deep_model = tc.logistic_classifier.create(image_train,target='label',features=['deep_features'])
```

### *Predicting some Data*

```
image_test[0:10]['image','label'].explore()
```

```
deep_model.predict(image_test[0:10])
```

<strong>So our deep model can recognise 9 images out of 10</strong>

### Evalute it's accuracy

```
deep_model.evaluate(image_test)
```

<strong>'accuracy': 0.79</strong>
