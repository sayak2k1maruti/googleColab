# ***Let's Code***

```
import turicreate  as tc
```

## *Loading test & train data*

```
image_train = tc.SFrame("image_train_data/")
image_test = tc.SFrame("image_test_data/")
```

### *Visualising some data*

```
image_train.explore()
```

## *Compute deep features from our images*

```
#deep_learning_model = tc.load_model('imagenet_model_iter45')
#image_data['deep_features'] = deep_learning_model.extract_features(image_data)
```

## *Create a nearest neighbors model to retrieve images from deep features*

```
model1 = tc.nearest_neighbors.create(image_train,
                                     features = ['deep_features'],
                                     label = 'id')
```

## *Make some predictions from our model*

```
image_train[20:25]['id','label' , 'image'].explore()
```

```
bird = image_train[24:25]
```

```
def get_images_from_ids(query_result):
  return image_train.filter_by(query_result['reference_label'],'id')
```

```
get_images_from_ids(model1.query(bird))['id','image','label'].explore()
```

### ***Defining a function to show all neighbour***

```
def show_neighbors(i):
  return get_images_from_ids(model1.query(image_train[i:i+1]))['image'].explore()
```

### ***Playing with it***

```
for i in range(1,100):
  show_neighbors(i)
```
